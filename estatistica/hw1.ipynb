```python
amostra_total_11 = np.array([0 for i in range(13)] + [1 for i in range(11)])
samples = np.random.choice(amostra_total_11, replace=True, size=(1000, len(amostra_total_11)))
mean = samples.sum(axis=1)/24
variance = np.array([m*(1-m) for m in list(mean)])
print('Esperança', variance.mean())
print('Variância', variance.std())
pd.Series(variance).plot(kind='hist', title='Distribuição da Variância para Distribuição de Bernoulli(p) com proporc. da Amostra 11/24')
```
